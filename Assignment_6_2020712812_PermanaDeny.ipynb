#Q1 Explain the difference between “MLP” and “CNN”

## Weight characteristic:
The MLP consists of multiple neural network, which each node is connected each other. The input in form of tabular data will is sent to each neuron on the layer and perform dot product between the weight and the input, therefore each node produce singgle value as output.


The CNN consists of neural network, which has convolution kernel. The input could be non-tabular data or even multidimension data. Each compnent-size of the input (pixel) is then convolved with a small matrix (called kernel) to produce output, and the same kernel moves into the next pixel depends on the stride parameter to produce another input. This procedure also known as weight or parameter sharing.

## Dataset
The MLP is suitable for tabular matrices data which each data does not always need to have a relationship (independent). therefore we can shuffle the input feature order if we need.

The CNN is suitable for non-tabular data, it can be a matrices, or tensor, which performs well to find local relationship (locality of pixel dependencies) and stationary of statistics. Therefore this model is typically eployed on image-type dataset.

## Implementation
The MLP can be used purely as MLP model, or to form a probability prediction or as a state of the art / SoTA in the last model layer.

The CNN can be used to perform picture analysis, and dimension reduction (coupled with pooling layer). usually, the CNN contains MLP as SoTA in the last layer to form probability prediction.
