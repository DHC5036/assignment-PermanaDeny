#Q1 Can you ever expect to see zero generalization error?

We would find zero generalization error when the y_predict and y_true exactly has the same value. 


in the training process, the training result might have zero generalization error when we only train small number of non complex dataset using large scale machine learning model. In other words, this zero generalization error is the strong indication of overfitting.



In the prediction or inference process using unseen dataset, it is almost impossible to get zero generalization error. The reason is that the unseen dataset obviously have different value from the training dataset. even though the accuracy is high, we typically get the generalization loss even with the small value.

# Q2 Why is dropout not typically used at test time?

Dropout makes our neural network model expect some degree of uncertaintly, thus the model learn better to generalize the feature from dataset.

In the general implementation, the dropout is deactivated during test time. It is becase we want the model generates a consistent result, besides we want the model occupy full capability of the network (since dropout disables some  node randomly inside the network).

However, we can enable dropout at the test time to examine the uncertainty of the model during the test time.
