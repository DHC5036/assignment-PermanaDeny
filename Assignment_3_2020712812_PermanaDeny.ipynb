It depends on what we are trying to classify. For example, suppose we build an MNIST prediction that has ten targets from 0 to 9. In this case, the most likely label will help us to decide specific label belongs to the MNIST image with certain confident level of probability.



However, when it comes to medical diagnosis, the most likely label typically leads to weak judgment. For instance, we design diabetes prediction with target positive (1) or negative (0). Supposedly the classifier says the patient has 0.22% diabetes positive. In other words, the classifier is 78% sure that the diagnosis is negative diabetes. The weak judgment comes by ignoring the small risk of 22% since the risk of diabetes is existed.
